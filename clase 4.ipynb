```python
def segundo_orden(y,t):
    return array ([y[1],-y[0]]) #es un vector
```

```python
def euler(func,p_ini,t_final,h): #vector
    pasos=arange (p_ini[0],t_final+h,h)  #funcion a integrar, su valor y en donde termina, h es pasos de integración
    y=[0]* len(pasos)
    y[0]=p_ini[1]
    for i in range(len(pasos)-1): # debido a que ya se había nombrado el punto inicial
        y[i+1]=y[i]+func(y[i],pasos[i])*h # lo que había antes mas las porciones nuevas
    sol=pasos,y # es la variable local
    return pasos,y
```

```python
sol=euler (segundo_orden,[0,[0,1]],200,0.001) # es la variable global, el promer elementoes un escalar y el segundo es un vector, hasta donde se quieren los pasos
y=zeros(len(sol[0])) # queremos que tenga la longitud de los pasos

for i in range(len(sol[1])):
    y[i]=sol[1][i][1] # i es en que renglon te situas, 1 que se toma el de la derecha
    
plot (sol[0],y) # numero de elementos de nuestro dominio contra el resultado que se obtuvo
plot(sol[0],cos(sol[0])) # elementos de nuestro dominio contra el seno
```
