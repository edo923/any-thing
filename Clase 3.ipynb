$$
f'(x) = \frac{f(x+h)-f(x)}{h}
$$

```python
def derivator (func,x,h):
    y = (func(x+h)-func(x))/h
    return y 
```

```python
plot(x,func(x))
plot(x,derivator(func,x,0.5))
plot(x,2*x)
```

# integracion por euler, lo que se queire integrar es:

$$
y'(x) = func(x)
$$

$$
y'(x)= y(x)
$$

$$
y(x) = A e^{x}
$$

```python
def mi_ec_dif(y,x):
    return y
```

$$
\int_{x_i}^{x_i + h} y'(x) \simeq y_{i} + h y(x_i)
$$

```python
def euler (func,p_ini,h,x_f):
    pasos = arange(p_ini[0],x_f+h,h)
    y = zeros(len(pasos))
    y[0] = p_ini[1]
    for i in range(len(pasos)-1):
        y[i+1] = y[i] + h*func(y[i], pasos[i])
    solucion = [pasos,y]
    return solucion
```

```python
sol = euler(mi_ec_dif,[0,1],0.01,200)
plot(sol[0],sol[1], sol[0],exp(sol[0]))
xlim (170,200)
# xlim delimita rangos 
```
